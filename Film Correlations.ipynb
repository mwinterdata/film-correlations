```python
# correlation matrix

correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix For Numeric Features')

plt.show()
```

```python
# numerize all categories

df_numerized = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

df_numerized
```

```python
#correlation matrix for all categories

correlation_matrix_all = df_numerized.corr()

sns.heatmap(correlation_matrix_all, annot = True)

plt.title('Correlation Matrix For All Categories')

plt.show()
```

```python
# unstack correlation matrix 

corr_pairs = correlation_matrix_all.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values(ascending = False)

sorted_pairs
```

```python
#other highly correlated factors with gross earnings

sorted_pairs_high = sorted_pairs[(sorted_pairs) > 0.5]

sorted_pairs_high = sorted_pairs_high[(sorted_pairs_high < 1)]

sorted_pairs_high_gross = sorted_pairs_high[['gross']]

sorted_pairs_high_gross 

# note: gross-name correlation should largely be ignored with most films having unique names and so providing almost no insight
```

### To conclude the highest correlating factor to gross revenue is the year a film was released, for which inflation is one obvious factor.

#### Going forward the next project shall be to adjust all gross earnings for inflation to determine whether this correlation still exists implying that individual films are also becoming on average more successful or whether this link is entirely due to inflation. 

#### This will also likely reduce writer, director, and star correlations to gross revenue as each individual will be be affected by their time working, although with careers generally lasting over a period of years/decades they will possibly be affected by inflation less than year released.

#### Another step to improve this project will be to focus on the distribution of gross revenue as due to the nature of film success most films earn very little while some earn huge amounts as demonstrated by the histogram below.
